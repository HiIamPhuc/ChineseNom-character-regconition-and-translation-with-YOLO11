```python
import torch

torch.cuda.empty_cache()

model = YOLO("yolo11n.pt")

model.train(data="/kaggle/input/ocrcharnom/data.yaml", epochs=300, batch=4, imgsz=640)
```

```python

model = YOLO(r"/content/best.pt")

res = model.predict(r"/content/DVSKTT-1 Quyen thu.DVSKTT_thu_I_1a.jpg")


for i in res:
    for box in i.boxes:
        print(box.xyxy, model.names[int(box.cls)])
    i.show()

```
